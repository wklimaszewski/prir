```
import tensorflow as tf
import numpy as np
from tensorflow import keras

def f(x):
  return x*x + 2*x

def metodaProstokatow(a, b, n):
    dx = (b - a) / n
    s = 0
    for i in range(1, n):
      s += f(a + i * dx)
    s *= dx
    return s

def metodaTrapezow(a, b, n):
    dx = (b - a) / n
    s = 0
    for i in range(1, n):
      s += f(a + i * dx)
    s = (s + (f(a) + f(b)) / 2) * dx
    return s

def metodaSimpsona(a, b, n):
  s = 0
  st = 0
  dx = (b - a) /n
  for i in range(1, n):
    x = a + i * dx
    st += f(x - dx / 2)
    if(i < n):
      s += f(x)
  s = dx / 6 * (f(a) + f(b) + 2 * s + 4 * st)
  return s

print('Wyniki dla metody prostokątów ', metodaProstokatow(0.0, 1.0, 1000))
print('Wyniki dla metody trapezów ', metodaTrapezow(0.0, 1.0, 1000))
print('Wyniki dla metody Simpsona ', metodaSimpsona(0.0, 1.0, 1000))
```

```
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([0.0, 8.0, 6.0, 0.0, -4.0, 0.0], dtype=float)
model.fit(xs, ys, epochs=500)
print(model.predict([10.0]))
```
